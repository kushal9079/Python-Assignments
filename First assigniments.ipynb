```python
# Question 1 :

# 1 string 
string = " Hello Pw"
print(string)

# 2 list 
list = " 10, 20, 30, 40 "
print(list)

# 3 float 
float = " 1.1, 2.2, 3.3, 4.4 "
print(float)

# 4 tuple
tuple = " 10, 20, 30, 40 "
print(tuple)

# Question 2 :

# var 1 is a empty string 

# var 2 is a String containing a list as a string

# var 3 is a list in form of string

# var 4 is a float value 


# Question 3 :

# / Division
print (10/2)
# This (/) operator is used to come Division of a value 

# % Modulus
print(5 % 2) 
# This (%) is mainly focus on remander part

# //  floor division
print (15 // 2)
# This (//)  floor division // rounds the result down to the nearest whole number

# ** Exponentiation
print(5 ** 2)
# This (**) Exponentiation or you can say , square of a number 


# Question 4 : 

my_list = [42, 'Hello', 3.14, True, [1, 2, 3], {'key': 'value'}, None, (4, 5, 6), 'World', False]

for element in my_list:
    print(f"Element: {element},         Data Type: {type  (element)    }")
    
    
# Question 5 :

# Input the values of A and B
A = int(input("Enter the value of A: "))
B = int(input("Enter the value of B: "))

# Initialize a variable to count the number of divisions
count = 0

# Use a while loop to check and count divisions
while A % B == 0:
    A = A // B  # Update the value of A after each division
    count += 1  # Increment the count

# Check if A is purely divisible by B
if count > 0:
    print(f"{A} is purely divisible by {B} and can be divided {count} times.")
else:
    print(f"{A} is not purely divisible by {B}.")
    
    
    
# Question 6 :

# Create a list of 25 integers
my_list = [2, 15, 8, 10, 27, 33, 45, 12, 18, 22, 7, 30, 41, 5, 13, 36, 20, 9, 25, 14, 6, 17, 11, 24, 19]

# Use a for loop to iterate through the list
for element in my_list:
    if element % 3 == 0:
        print("{element} is divisible by 3.")
    else:
        print("{element} is not divisible by 3.")
        
        
        
# Question  7  :

# In Python, the terms "mutable" and "immutable" refer to whether an object's state can be modified after it is created.
# Mutable Data Types:

# Mutable objects can be altered or modified after creation.
# Examples of mutable data types in Python include lists, dictionaries, and sets.
# When you modify a mutable object, the changes affect the original object.


# Example of a mutable type: list

mutable_list = [1, 2, 3]
print("Original List:", mutable_list)

# Modifying the list

mutable_list[0] = 10
print("Modified List:", mutable_list)


#Immutable Data Types:

#Immutable objects cannot be altered or modified after creation.
#Examples of immutable data types in Python include integers, floats, strings, and tuples.
#Any operation that appears to modify an immutable object actually creates a new object.


# Example of an immutable type: string

immutable_string = "Hello"
print("Original String:", immutable_string)

# Concatenating to create a new string

new_string = immutable_string + " World"
print("New String:", new_string)
print("Original String (unchanged):", immutable_string)





```
